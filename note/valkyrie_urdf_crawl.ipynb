```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('/home/jacob/Documents/PersonaIsaacLab/external/valkyrie/src/main/resources/models/val_description/urdf/valkyrie_sim.urdf')
```

```python
root = tree.getroot()
```

```python
joints = {}

for child in root:
    if child.tag == "joint":
        print(child.attrib)
        joints[child.attrib['name']] = {}
        for child2 in child:
            if child2.tag in ['limit', 'controller_gains', 'dynamics']:
                for attrib in child2.attrib:
                    joints[child.attrib['name']][f'{child2.tag}_{attrib}'] = child2.attrib[attrib]
                print(child2.tag)
                print(child2.attrib)

                
```

```python
# keep joint only if there are children 
joints = {k:v for k,v in joints.items() if v}
```

```python
joints
```

```python

output = []

for joint_name, joint_values in joints.items():
    params = [
        f'joint_names_expr=["{joint_name}"]'
    ]
    if "limit_effort" in joint_values:
        params.append(f"effort_limit={float(joint_values['limit_effort'])}")
    if "limit_velocity" in joint_values:
        params.append(f"velocity_limit={float(joint_values['limit_velocity'])}")
    if "controller_gains_Kp" in joint_values:
        params.append(f"stiffness={float(joint_values['controller_gains_Kp'])}")
    else:
        params.append("stiffness=0.0")
    if "dynamics_damping" in joint_values and float(joint_values["dynamics_damping"]) != 0.0:
        params.append(f"damping={float(joint_values['dynamics_damping'])}")
    if "dynamics_friction" in joint_values and float(joint_values["dynamics_friction"]) != 0.0:
        params.append(f"friction={float(joint_values['dynamics_friction'])}")

    # Join parameters into the actuator config
    param_str = ",\n        ".join(params)
    config = f'"{joint_name}": ImplicitActuatorCfg(\n        {param_str}\n    ),'
    output.append(config)

# Print generated Python code
print("actuators = {\n" + "\n".join(output) + "\n}")
```
