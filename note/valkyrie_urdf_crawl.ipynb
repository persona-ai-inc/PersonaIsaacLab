```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('/home/jacob/Documents/PersonaIsaacLab/external/valkyrie/src/main/resources/models/val_description/urdf/valkyrie_sim.urdf')
```

```python
root = tree.getroot()
```

```python
joints = {}

for child in root:
    if child.tag == "joint":
        print(child.attrib)
        joints[child.attrib['name']] = {}
        for child2 in child:
            if child2.tag in ['limit', 'controller_gains', 'dynamics']:
                for attrib in child2.attrib:
                    joints[child.attrib['name']][f'{child2.tag}_{attrib}'] = child2.attrib[attrib]
                print(child2.tag)
                print(child2.attrib)

                
```

```python
# keep joint only if there are children 
joints = {k:v for k,v in joints.items() if v}
```
